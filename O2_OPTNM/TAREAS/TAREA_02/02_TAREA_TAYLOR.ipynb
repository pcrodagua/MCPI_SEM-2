```python
import math
import numpy as np
import sympy as sy
import matplotlib.pyplot as plt
from sympy.functions import sin, cos
%matplotlib inline
```

$$f(x) = \sum _{n = 0}^{\infty}{\frac{f^{(n)}(x_0)}{n!}}(x - x_0)^n$$

```python
def taylor(function, x_0, n):
    p = 0
    for i in range(0, n + 1):
        p += (function.diff(x, i).subs(x, x_0))/(math.factorial(i))*(x - x_0)**i
    return p
```

```python
def plot(func, orden):
    x_lims = [-2*np.pi, 2*np.pi]
    x_1 = np.linspace(x_lims[0],x_lims[1],100)
    y_1 = []
    for j in range(1,orden,1):
        tfunc = taylor(func,0,j)
        print('n = {0} {1}'.format(str(j), tfunc))
        for k in x_1:
            y_1.append(tfunc.subs(x, k))
        plt.plot(x_1, y_1, label='order '+str(j))
        y_1 = []
    plt.plot(x_1,np.sin(x_1),label='sin(x)')
    plt.xlim(x_lims)
    plt.ylim([-10, 10])
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.title('Taylor Polynomial Series')
    plt.show()
```

```python
x = sy.Symbol('x')
f = sin(x)
orden=10
plot(f,orden)
```
