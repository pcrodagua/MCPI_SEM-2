```python
# A non-local algorithm for noise filtering
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils as im
from skimage import util as sku
```

```python
input_img = cv.imread('im_contrast_5.jpg', 1)
plt.imshow(input_img, cmap='gray', vmin=0, vmax=255)
sp_img = sku.random_noise(input_img, 'gaussian')*255
sp_img = np.uint8(sp_img)
```
