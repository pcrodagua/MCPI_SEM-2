```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils as im
```

```python
def size_of_img(image):
    row, col = image.shape
    print(row, col)
```

```python
input_img = cv.imread('im_contrast_5.jpg', 0)
plt.imshow(input_img, cmap='gray', vmin=0, vmax=255)
size_of_img(input_img)
```

```python
def create_border(img):
    borderType = cv.BORDER_CONSTANT
    TDLU=[1, 1, 1, 1] 
    img = cv.copyMakeBorder(img, TDLU[0] , TDLU[1] , TDLU[2] , TDLU[3] , borderType, None, 255)
    row, col = img.shape
    return img
```

```python
output_img = create_border(input_img)
plt.imshow(output_img, cmap='gray', vmin=0, vmax=255)
size_of_img(output_img)
```

```python
kernel = np.ones((3,3),np.float32)/9
processed_image = cv.filter2D(output_img,-1,kernel)
```

```python
def mean_filter(image):
    w = 2
    img = image
    for i in range(2, image.shape[0]-2):
        for j in range(2, image.shape[1]-2):
            block = image[i - w:i + w + 1, j - w:j + w +1]
            m = np.mean(block, dtype = np.float32)
            img[i][j] = int(m)
    return img
```

```python
plt.imshow(processed_image, cmap='gray', vmin=0, vmax=255)
size_of_img(processed_image)
```

```python
processed_image2 = mean_filter(output_img)
```

```python
plt.imshow(processed_image2, cmap='gray', vmin=0, vmax=255)
size_of_img(processed_image2)
```

```python
def remove_border(img):
    endr, endc = processed_image2.shape[0:2] 
    img = np.delete(img, endr-1, axis=0)
    img = np.delete(img, endc-1, axis=1)
    
    img = np.delete(img, 0, axis=0)
    img = np.delete(img, 0, axis=1)
    return img
```

```python
processed_image3 = remove_border(processed_image2)
```

```python
plt.imshow(processed_image3, cmap='gray', vmin=0, vmax=255)
size_of_img(processed_image3)
```
