```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import imutils as im
from skimage import util as sku
```

```python
input_img = cv.imread('im_contrast_5.jpg', 1)
plt.imshow(input_img, cmap='gray', vmin=0, vmax=255)
sp_img = sku.random_noise(input_img, 'salt')*255
sp_img = np.uint8(sp_img)
```

```python
plt.imshow(sp_img, cmap='gray', vmin=0, vmax=255)
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
import cv2 
# With jupyter notebook uncomment below line 
%matplotlib inline 
# This plots figures inside the notebook

def plot_cv_img(input_image, output_image1):     
    """     
    Converts an image from BGR to RGB and plots     
    """   

    fig, ax = plt.subplots(nrows=1, ncols=2)

    ax[0].imshow(cv2.cvtColor(input_image, cv2.COLOR_BGR2RGB))          
    ax[0].set_title('Input Image')
    ax[0].axis('off')
    
    ax[1].imshow(cv2.cvtColor(output_image1, cv2.COLOR_BGR2RGB))          
    ax[1].set_title('Median Filter (3,3)')
    ax[1].axis('off')

    
    # plt.savefig('../figures/03_median_filter.png')

    plt.show()
    

def main():
    # read an image 
    img = cv2.imread('im_contrast_5.jpg')
    
    
    median1 = cv2.medianBlur(sp_img,3)
    #median2 = cv2.medianBlur(sp_img,5)
    #median3 = cv2.medianBlur(sp_img,7)
    
    
    # Do plot
    plot_cv_img(sp_img, median1)

if __name__ == '__main__':
    main()
```
